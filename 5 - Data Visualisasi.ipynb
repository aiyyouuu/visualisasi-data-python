```python
# import packages
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn

# setting plotting style as global parameters
plt.rc('figure', figsize=(10, 5))  # figure size
seaborn.set()  # use seaborn style
```

```python
# generate dummy data
x = np.linspace(-2, 2, 5)

y1 = x**2  # quadratic function
y2 = x # line farunction

plt.plot(x, y1, label='quadratic')
plt.plot(x, y2, label='linear')

plt.legend(loc='best')
plt.title('Linear Function vs Quadratic Function', fontsize=18)
plt.xlabel('x', fontsize=15)
plt.ylabel('y',fontsize=15)
plt.show()
```

```python
x = np.linspace(0, 5 * np.pi, 1000)
y = np.cos(x)

plt.plot(x, y, label='cos function')

plt.title('Cos Line Function', fontsize=18)
plt.xlabel('x')
plt.ylabel('y');
plt.show()
```

```python
data = np.random.normal([1, 1], [3, 2], (500, 2))

plt.scatter(data[:, 0], data[:, 1], alpha=0.8);
plt.title('Scatter Plots', fontsize=18)
plt.show()
```

```python
x = np.linspace(0, 5 * np.pi, 1000)
y = np.cos(x)
noisy_y = np.random.normal(np.cos(x), 0.1)

plt.plot(x, y, color='w', linewidth=2, label='y = cos function')
plt.scatter(x, noisy_y, alpha=0.5, label='noisy y');
plt.legend(loc='best');

plt.title('Cos Line and Scatter Plots', fontsize=18)
plt.xlabel('x')
plt.ylabel('y');
plt.show()
```

```python
samples = np.random.normal(1.0, 1, size=1000)  # sample from normal distribution

print(samples.shape)
print(samples.dtype)
print(samples[:30])

plt.hist(samples, bins=30, linestyle='-');
plt.title('Histogram', fontsize=18)
plt.show()
```

```python
x = np.arange(10)
y = np.random.randint(1, 5, 10)
plt.bar(x, y)
plt.title('Bar Plot', fontsize=18)
plt.show()
```
